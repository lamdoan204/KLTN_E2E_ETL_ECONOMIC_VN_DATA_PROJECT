```python
import pandas as pd 

file_path ='../../../Bronze_Excel_Financial_Report/2016/Is_Quy/Bieu-6.2016.xlsx'
excel_file = pd.ExcelFile(file_path)
all_sheet = excel_file.sheet_names
all_sheet
```

```python
import unicodedata
import re
def clean_text(s):
    # bỏ dấu tiếng Việt
    s = unicodedata.normalize('NFD', s)
    s = ''.join(c for c in s if unicodedata.category(c) != 'Mn')
    
    # giữ lại chữ cái
    s = re.sub(r'[^a-zA-Z]', '', s)
    
    return s.lower()

s = "Tổng sản phẩm trong nước (GDP)!!!"
print(clean_text(s))
```

```python
#  Kiểm tra có công bố VDTTXH không
vdttxh_sheet_index = -1
for i in range(len(all_sheet)):
    sheet_name = all_sheet[i]
    cleaned_sheet_name = clean_text(sheet_name)
    if 'ttxh' in cleaned_sheet_name:
        vdttxh_sheet_index = i
        break

if(vdttxh_sheet_index == -1):
    print("Khong cong bo vddtxh")
else:
    print(vdttxh_sheet_index)
```

```python
vdt_sheet = pd.read_excel(file_path, sheet_name=all_sheet[vdttxh_sheet_index], header= None)
```

```python
column_names = ['investment_type', 'thuc_hien_quy_truoc', 'uoc_tinh_quy_nay']
vdt_sheet = vdt_sheet.iloc[::, 1:4]
vdt_sheet.columns = column_names
```

```python
num_of_removed_row = -1
col = vdt_sheet['uoc_tinh_quy_nay']
for i in range(len(col)):
    num_of_removed_row += 1
    if type(col[i]) is str:
        break

vdt_sheet = vdt_sheet.iloc[num_of_removed_row::, ::].reset_index(drop= True)
```
